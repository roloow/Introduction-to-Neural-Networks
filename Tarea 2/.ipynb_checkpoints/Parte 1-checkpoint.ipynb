## A)

Se define una función para leer los archivos batches y se utiliza el *encoding* **Latin1** para resolver problemas de lectura *'asci'*. **NOTAR** el uso de la librería *pickle*, dado que en python 3.5 cPickle es reducido a una librería *_pickle* y subutilizada por *pickle*. [StackOverFlow #Issue](https://stackoverflow.com/questions/37132899/installing-cpickle-with-python-3-5)

````python
def unpickle(file):
    fo = open(file, 'rb')
    dict = pickle.load(fo, encoding='latin1')
    fo.close()
    return dict
````

Se procede a continuar la ejecución leyendo los datos y guardandolos en variables.

````python
data1 = unpickle("Scripts/Parte1/data_batch_1")
data2 = unpickle("Scripts/Parte1/data_batch_2")
data3 = unpickle("Scripts/Parte1/data_batch_3")
data4 = unpickle("Scripts/Parte1/data_batch_4")
data5 = unpickle("Scripts/Parte1/data_batch_5")
test_batck = unpickle("Scripts/Parte1/test_batch")
````

Luego de cargar los diccionarios se procede a generar funciones para crear los conjuntos de entrenamiento y test. Esto se hace creando dos funcionalidades de lectura de archivos.

````python
def load_CIFAR_one(filename):
    with open(filename, 'rb') as f:
        datadict = pickle.load(f, encoding='latin1')
        X = datadict['data']
        Y = datadict['labels']
        Y = np.array(Y)
        return X, Y

def load_CIFAR10(PATH):
    xs = []
    ys = []
    for b in range(1,6):
        f = os.path.join(PATH, 'data_batch_%d' % (b, ))
        X, Y = load_CIFAR_one(f)
        xs.append(X)
        ys.append(Y)
    Xtr = np.concatenate(xs)
    Ytr = np.concatenate(ys)
    del X, Y
    Xte, Yte = load_CIFAR_one(os.path.join(PATH, 'test_batch'))
    return Xtr, Ytr, Xte, Yte
````

Ya creadas las funciones, se procede a crear los conjuntos de entrenamiento y de testing, para luego, en base a los nombres de categorías usados en la tarea 1, señalados como **class_names**. La imagenes son redimensionadas y traspuestas para poder verse. La imagenes mostradas equivalen a la clasificación esperable de la red.

````python
x_train, y_train, x_test, y_test = load_CIFAR10('Scripts/Parte1/')

class_names = ['Avion', 'Auto', 'Ave', 'Gato', 'Venado', 'Perro', 'Rana', 'Caballo', 'Barco', 'Camion']

PLOT_IMAGES = x_test.reshape((x_test.shape[0],32,32,3), order="F").transpose([0, 2, 1, 3])
fig = plt.figure(figsize=(10, 10))
for i in range(36):
    fig.add_subplot(6, 6, (i+1), xticks=[], yticks=[])
    plt.title("{0}".format(class_names[list(y_test)[i]]))
    plt.imshow(PLOT_IMAGES[i])
plt.show()
````

Luego, para poder continuar con las labores, se transforman todos los conjuntos según las necesidades de tensorflow.

````python
x_train = x_train.reshape((x_train.shape[0],32,32,3), order="F").transpose([0, 2, 1, 3])
x_test= x_test.reshape((x_test.shape[0],32,32,3), order="F").transpose([0, 2, 1, 3])

num_classes = 10

y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

print ("Formato de entrenamiento:", x_train.shape)
print ("Formato de testeo:", x_test.shape)
````
